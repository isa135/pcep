Vamos agora brincar com triângulos. Vamos começar com uma função para verificar se três lados de determinados comprimentos podem construir um triângulo.
    
    Let's play with triangles now. We'll start with a function to check whether three sides of given lengths can build a triangle.

Sabemos da escola que a soma de dois lados arbitrários tem de ser mais longa do que o terceiro lado.
    
    We know from school that the sum of two arbitrary sides has to be longer than the third side.

Não será um desafio difícil. A função terá três parâmetros - um para cada lado.
    
    It won't be a hard challenge. The function will have three parameters - one for each side.

Ela vai devolver True se os lados puderem construir um triângulo, e False caso contrário. Neste caso, is_a_triangle é um bom nome para tal função.
    
    It will return True if the sides can build a triangle, and False otherwise. In this case, is_a_triangle is a good name for such a function.

Veja o código no editor. Pode encontrar aí a nossa função. Execute o programa.
    
    Look at the code in the editor. You can find our function there. Run the program.

Parece que funciona bem - estes são os resultados:
    
    It seems that it works well - these are the results:

```python
def is_a_triangle(a, b, c):
    if a + b <= c:
        return False
    if b + c <= a:
        return False
    if c + a <= b:
        return False
    return True


print(is_a_triangle(1, 1, 1))
print(is_a_triangle(1, 1, 3))
```

Podemos torná-lo mais compacto? Parece um pouco palavroso.
    
    Can we make it more compact? It looks a bit wordy.

Esta é uma versão mais compacta:
    
    This is a more compact version:

```python
def is_a_triangle(a, b, c):
    if a + b <= c or b + c <= a or c + a <= b:
        return False
    return True
print(is_a_triangle(1, 1, 1))
print(is_a_triangle(1, 1, 3))
```

Podemos compactá-lo ainda mais?
    
    Can we compact it even more?

Sim, podemos - veja:
    
    Yes, we can - look:

```python
def is_a_triangle(a, b, c):
    return a + b > c and b + c > a and c + a > b
print(is_a_triangle(1, 1, 1))
print(is_a_triangle(1, 1, 3))
```

Negamos a condição (invertemos os operadores relacionais e substituímos orpor and, recebendo uma expressão universal para testar triângulos).
    
    We've negated the condition (reversed the relational operators and replaced ors with ands, receiving a universal expression for testing triangles).
