##### Como funcionam as funções
    
    How functions work

    Look at the picture below:

Ela tenta mostrar-lhe todo o processo:

    It tries to show you the whole process:
    
•	quando se invoca uma função, o Python lembra-se do local onde aconteceu e salta para a função invocada;

    •	when you invoke a function, Python remembers the place where it happened and jumps into the invoked function;
    
•	o corpo da função é então executado;

    •	the body of the function is then executed;
    
•	chegar ao fim da função força o Python a regressar ao local diretamente após o ponto de invocação.

    •	reaching the end of the function forces Python to return to the place directly after the point of invocation.
    
Há dois, muito importantes, senão. Aqui está o primeiro deles:
    
    There are two, very important, catches. Here's the first of them:

Não se deve invocar uma função que não seja conhecida no momento da invocação.
    
    You mustn't invoke a function which is not known at the moment of invocation.

Lembre-se - o Python lê o seu código de cima para baixo. Não vai olhar em frente para encontrar uma função que se esqueceu de colocar no lugar certo ("certo" significa "antes da invocação").
    
    Remember - Python reads your code from top to bottom. It's not going to look ahead in order to find a function you forgot to put in the right place ("right" means "before invocation".)
    
Inserimos um erro neste código - consegue ver a diferença?
    
    We've inserted an error into this code - can you see the difference?

```python
print("We start here.")
message()
print("We end here.")

def message():
    print("Enter a value: ")
```

Movemos a função para o final do código. O Python é capaz de encontrá-la quando a execução atinge a invocação?
    
    We've moved the function to the end of the code. Is Python able to find it when the execution reaches the invocation?

Não, não é. A mensagem de erro lerá:
    
    No, it isn't. The error message will read:



Não tente forçar o Python a procurar por funções que não entregou na altura certa.

    Don't try to force Python to look for functions you didn't deliver at the right time.
    
O segundo senão soa um pouco mais simples:
    
    The second catch sounds a little simpler:

Não se deve ter uma função e uma variável com o mesmo nome.
    
    You mustn't have a function and a variable of the same name.

O seguinte snippet está errado:
    
    The following snippet is erroneous:

```python
def message():
    print("Enter a value: ")

message = 1
```

Atribuir um valor à mensagem do nome faz com que o Python esqueça o seu papel anterior. A função chamada message torna-se indisponível.
    
    Assigning a value to the name message causes Python to forget its previous role. The function named message becomes unavailable.
    
Felizmente, é livre de misturar o seu código com funções - não é obrigado a colocar todas as suas funções no topo do seu source file.
    
    Fortunately, you're free to mix your code with functions - you're not obliged to put all your functions at the top of your source file.

Veja o snippet:
    
    Look at the snippet:

```python
print("We start here.")

def message():
    print("Enter a value: ")

message()
print("We end here.")
```

Pode parecer estranho, mas está completamente correto, e funciona como pretendido.
    
    It may look strange, but it's completely correct, and works as intended.
Voltemos ao nosso exemplo principal, e utilizemos a função para o trabalho certo, como aqui:
    
    Let's return to our primary example, and employ the function for the right job, like here:

```python
def message():
    print("Enter a value: ")

message()
a = int(input())
message()
b = int(input())
message()
c = int(input())
```

Modificar a mensagem de solicitação é agora fácil e claro - pode fazê-lo alterando o código num só lugar - dentro do corpo da função.
    
    Modifying the prompting message is now easy and clear - you can do it by changing the code in just one place - inside the function's body.
