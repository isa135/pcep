Vamos começar com uma função para avaliar o Índice de Massa Corporal (IMC).
    
    Let's get started on a function to evaluate the Body Mass Index (BMI).

Como pode ver, a fórmula obtém dois valores:

    As you can see, the formula gets two values:
    
•	peso (originalmente em quilogramas)
    
    •	weight (originally in kilograms)
    
•	altura (originalmente em metros)

    •	height (originally in meters)

Parece que esta nova função terá dois parâmetros. O seu nome será bmi, mas se preferir qualquer outro nome, use-o.
    
    It seems that this new function will have two parameters. Its name will be bmi, but if you prefer any other name, use it instead.

Vamos codificar a função:
    
    Let's code the function:

```python
def bmi(weight, height):
    return weight / height ** 2

print(bmi(52.5, 1.65))
```

A função cumpre as nossas expetativas, mas é um pouco simples - assume que os valores de ambos os parâmetros são sempre significativos. Vale definitivamente a pena verificar se são dignos de confiança.
    
    The function fulfils our expectations, but it's a bit simple - it assumes that the values of both parameters are always meaningful. It's definitely worth checking if they're trustworthy.
