##### Algumas funções simples: avaliar a área de um triângulo
    
    Some simple functions: evaluating a triangle's area

Também podemos avaliar a área de um triângulo. A fórmula de Heron será útil aqui:
    
    We can also evaluate a triangle's area. Heron's formula will be handy here:

Vamos usar o operador de exponenciação para encontrar a raiz quadrada - pode parecer estranho, mas funciona:
    
    We're going use the exponentiation operator to find the square root - it may seem strange, but it works:

Este é o código resultante:
    
    This is the resulting code:

```python
def is_a_triangle(a, b, c):
    return a + b > c and b + c > a and c + a > b
def heron(a, b, c):
    p = (a + b + c) / 2
    return (p * (p - a) * (p - b) * (p - c)) ** 0.5
def area_of_triangle(a, b, c):
    if not is_a_triangle(a, b, c):
        return None
    return heron(a, b, c)
print(area_of_triangle(1., 1., 2. ** .5))
```
