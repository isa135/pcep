Funções parametrizadas: continuação
    
    Parametrized functions: continued

Uma função pode ter tantos parâmetros quantos quiser, mas quanto mais parâmetros tiver, mais difícil é memorizar os seus papéis e propósitos.
    
    A function can have as many parameters as you want, but the more parameters you have, the harder it is to memorize their roles and purposes.

Vamos modificar a função - tem dois parâmetros agora:
    
    Let's modify the function - it has two parameters now:

```python
def message(what, number):
    print("Enter", what, "number", number)

# invoke the function
```

Isto também significa que invocar a função irá requerer dois argumentos.
    
    This also means that invoking the function will require two arguments.

O primeiro parâmetro novo destina-se a levar o nome do valor desejado.
    
    The first new parameter is intended to carry the name of the desired value.

Aqui está:
    
    Here it is:

```python
def message(what, number):
    print("Enter", what, "number", number)

message("telephone", 11)
message("price", 5)
message("number", "number")
```
