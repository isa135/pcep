##### O que é um dicionário?
    
    What is a dictionary?

O dicionário é outra estrutura de dados Python. Não é um tipo de sequência (mas pode ser facilmente adaptado ao processamento de sequências) e é mutável.
    
    The dictionary is another Python data structure. It's not a sequence type (but can be easily adapted to sequence processing) and it is mutable.

Para explicar o que é realmente o dicionário Python, é importante compreender que se trata literalmente de um dicionário.
    
    To explain what the Python dictionary actually is, it is important to understand that it is literally a dictionary.

O dicionário Python funciona da mesma forma que um dicionário bilingue. Por exemplo, tem uma palavra inglesa (por exemplo, cat) e precisa do seu equivalente francês. Navega no dicionário para encontrar a palavra (pode usar diferentes técnicas para o fazer - não importa) e eventualmente obtém-na. A seguir, verifica o homólogo francês e é (muito provavelmente) a palavra "chat".
    
    The Python dictionary works in the same way as a bilingual dictionary. For example, you have an English word (e.g., cat) and need its French equivalent. You browse the dictionary in order to find the word (you may use different techniques to do that - it doesn't matter) and eventually you get it. Next, you check the French counterpart and it is (most probably) the word "chat".

No mundo de Python, a palavra que se procura chama-se key. A palavra que obtém do dicionário chama-se um value.
    
    In Python's world, the word you look for is named a key. The word you get from the dictionary is called a value.

Isto significa que um dicionário é um conjunto de pares de key-values (valores-chave). Nota:
    
    This means that a dictionary is a set of key-value pairs. Note:
    
•	cada chave deve ser única - não é possível ter mais do que uma chave com o mesmo valor;
    
    •	each key must be unique - it's not possible to have more than one key of the same value;
    
•	uma chave pode ser qualquer tipo de objeto imutável: pode ser um número (inteiro ou float), ou mesmo uma string, mas não uma lista;

    •	a key may be any immutable type of object: it can be a number (integer or float), or even a string, but not a list;
    
•	um dicionário não é uma lista - uma lista contém um conjunto de valores numerados, enquanto que um dicionário contém pares de valores;

    •	a dictionary is not a list - a list contains a set of numbered values, while a dictionary holds pairs of values;
    
•	a função len() funciona também para dicionários - devolve o número de elementos de key-value no dicionário;

    •	the len() function works for dictionaries, too - it returns the numbers of key-value elements in the dictionary;
    
•	um dicionário é uma ferramenta de sentido único - se tiver um dicionário inglês-francês, pode procurar por equivalentes franceses de termos ingleses, mas não vice-versa.

    •	a dictionary is a one-way tool - if you have an English-French dictionary, you can look for French equivalents of English terms, but not vice versa.

##### Como fazer um dicionário?

    How to make a dictionary?

Se quiser atribuir alguns pares iniciais a um dicionário, deverá utilizar a seguinte sintaxe:
    
    If you want to assign some initial pairs to a dictionary, you should use the following syntax:

```python
dictionary = {"dog": "chien", "cat": "chat", "horse": "cheval"}
phone_numbers = {'boss': 5551234567, 'Suzy': 22657854310}
empty_dictionary = {}

print(dictionary)
print(phone_numbers)
print(empty_dictionary)
```

No primeiro exemplo, o dicionário usa chaves e valores que são ambas strings. No segundo, as chaves são strings, mas os valores são inteiros. A disposição inversa (teclas → números, valores → strings) também é possível, bem como a combinação número-número.
    
    In the first example, the dictionary uses keys and values which are both strings. In the second one, the keys are strings, but the values are integers. The reverse layout (keys → numbers, values → strings) is also possible, as well as number-number combination.

A lista de pares é rodeada por chavetas, enquanto os pares em si são separados por vírgulas, e as chaves e valores por dois pontos.
    
    The list of pairs is surrounded by curly braces, while the pairs themselves are separated by commas, and the keys and values by colons.

O primeiro dos nossos dicionários é um dicionário inglês-francês muito simples. O segundo - uma diretoria telefónica muito pequena.
    
    The first of our dictionaries is a very simple English-French dictionary. The second - a very tiny telephone directory.

Os dicionários vazios são construídos por um par vazio de chavetas - nada de invulgar.
    
    The empty dictionaries are constructed by an empty pair of curly braces - nothing unusual.

Notou alguma coisa surpreendente? A ordem dos pares impressos é diferente do que na atribuição inicial. O que é que isso significa?
    
    Have you noticed anything surprising? The order of the printed pairs is different than in the initial assignment. What does that mean?

Em primeiro lugar, é uma confirmação de que os dicionários não são listas - não preservam a ordem dos seus dados, uma vez que a ordem não tem qualquer significado (ao contrário do que acontece nos dicionários de papel reais). A ordem em que um dicionário armazena os seus dados está completamente fora do seu controlo, e das suas expectativas. Isso é normal. (*)
    
    First of all, it's a confirmation that dictionaries are not lists - they don't preserve the order of their data, as the order is completely meaningless (unlike in real, paper dictionaries). The order in which a dictionary stores its data is completely out of your control, and your expectations. That's normal. (*)

NOTA

(*) Em Python 3.6x os dicionários tornaram-se coleções ordenadas por padrão. Os seus resultados podem variar dependendo da versão Python que estiver a utilizar.
    
    (*) In Python 3.6x dictionaries have become ordered collections by default. Your results may vary depending on what Python version you're using.
