##### Os loops for e o ramo else .
    
    The for loop and the else branch

for os loops comportam-se de forma um pouco diferente - dê uma vista de olhos ao snippet no editor e execute-o.
    
    for loops behave a bit differently - take a look at the snippet in the editor and run it.

A variável i retém o seu último valor.
    
    The i variable retains its last value.

Modifique um pouco o código para levar a cabo mais uma experiência.
    
    Modify the code a bit to carry out one more experiment.

```python
i = 111
for i in range(2, 100):
    print(i)
else:
    print("else:", i)
```

O corpo do loop não será executado aqui. Nota: atribuímos a variável i antes do loop.

    The loop's body won't be executed here at all. Note: we've assigned the i variable before the loop.

Quando o corpo do loop não é executado, a variável de controlo retém o valor que tinha antes do loop.
    
    When the loop's body isn't executed, the control variable retains the value it had before the loop.

Nota: se a variável de controlo não existir antes do início do loop, não existirá quando a execução atingir o ramo else .
    
    Note: if the control variable doesn't exist before the loop starts, it won't exist when the execution reaches the else branch.
