##### Listas em listas: arrays bidimensionais – continuação
    
    Lists in lists: two-dimensional arrays - continued

O acesso ao campo selecionado do tabuleiro requer dois índices - o primeiro seleciona a linha; o segundo - o número do campo dentro da linha, que é de fato um número de coluna.
    
    Access to the selected field of the board requires two indices - the first selects the row; the second - the field number inside the row, which is de facto a column number.

Dê uma vista de olhos no tabuleiro de xadrez. Cada campo contém um par de índices que devem ser dados para aceder ao conteúdo do campo:
    
    Take a look at the chessboard. Every field contains a pair of indices which should be given to access the field's content:

```python
EMPTY = "-"
ROOK = "ROOK"
board = []

for i in range(8):
    row = [EMPTY for i in range(8)]
    board.append(row)
    
print(board)
```

Olhando para a figura mostrada acima, vamos colocar algumas peças de xadrez no tabuleiro. Primeiro, vamos adicionar todas as torres (em inglês, rooks):
    
    Glancing at the figure shown above, let's set some chess pieces on the board. First, let's add all the rooks:

```python
board[0][0] = ROOK
board[0][7] = ROOK
board[7][0] = ROOK
board[7][7] = ROOK

print(board)
```

Se quiser acrescentar um cavaleiro (knight) ao C4, faça-o da seguinte forma:
    
    If you want to add a knight to C4, you do it as follows:

```python
board[4][2] = "KNIGHT"
print(board)
```

E agora um peão (pawn) para E5:
    
    And now a pawn to E5:

```python
board[3][4] = "PAWN"
print(board)
```
